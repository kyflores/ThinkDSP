2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

To find the window corresponding to a filter, we can create a spectrum that represents the response of the filter, and then look at the wave from that spectrum. This is valid because convolution with the window is the same as multiplication of the signal's DFT with the window's spectrum.  First, I'll try to make the ideal high pass filter: 0 below the pass band, unity above. I'll test its behavior against a sawtooth wave.

```python
hp=np.concatenate((np.zeros(400),np.ones(5113))) #5113 to balance the length of the sawwave spectrum to us elementwise mult later.
hp_spec=thinkdsp.Spectrum(hp,11025)

lp=np.concatenate((np.ones(400),np.zeros(5113)))
lp_spec=thinkdsp.Spectrum(lp,11025)

sawwave=thinkdsp.SawtoothSignal(100).make_wave()
sawwave.make_spectrum().plot(high=1000) #Unmodified sawtooth.
print(len(sawwave.make_spectrum().hs))
```

```python
hp_spec.plot(high=1000) #Filter
```

```python
sawtooth_hp=sawwave.make_spectrum()*hp_spec
sawtooth_hp.plot(high=1000)
```

Now that we know our filter produces the desired behavior, let's see what kind of the window shape the filter produces.

```python
w=hp_spec.make_wave().plot()
#High pass filter w/ cutoff at 400hz
```

```python
w=lp_spec.make_wave().plot()
#Low pass filter w/ cutoff at 400hz
```

These window functions have a very unusual wave shape and aren't easily described by any function I know off. Their amplitude being very small throughout is a reasonable outcome: Because they contain so many frequencies, they appear to experience mostly destructive interference throughout, except for at the ends.

3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
#Using the code from Chap07

boxcar = np.ones(100)
boxcar /= sum(boxcar)
bc=thinkdsp.Wave(boxcar,11025)
bcspec=bc.make_spectrum()
bcspec.plot()
js1=np.angle(bcspec.hs)

gaussian = scipy.signal.gaussian(M=100, std=2)
gaussian /= sum(gaussian)
gs=thinkdsp.Wave(gaussian,11025)
gsspec=gs.make_spectrum()
gsspec.plot()
js2=np.angle(gsspec.hs)
```

Above are the spectra of the boxcar and the gaussian. The box is completely flat so its only frequency should be zero, and the gaussian's spectrum is known to also be a gaussian.

```python
thinkplot.plot(js1,label='Boxcar')
thinkplot.plot(js2,label='Gaussian')
thinkplot.config(xlabel='frequency',
                 ylabel='phase angle',
                 loc='best')
```

The boxcar window's phase appears to be pretty much 0 (those bumps corresponding to single points that aren't 0) while the gaussian phase angle oscillates against an upward trend. I can't think of a good reason why this occurs.

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

First, it'll be kind of difficult to see anything from the spectrum of two points so I'll try to show some signal convolved against each window. If the first derivative is like multiplication of each magnitude by $f$, I might reasonably predict the second derivative to be like multiplication of magnitudes by $f^2$. The intuition behind this would be that derivatives express the rate of change of a signal, and because signals with a high frequency change quickly, they are more similar to the first derivative function.

Since a second derivative is the result of taking the derivative twice in succession, you might expect it to stack the same way as sticking two high pass filters in series: the cutoff frequency is the same, but the rolloff is more severe.

```python
signal=thinkdsp.SawtoothSignal(100).make_wave()

f_d=np.array([1,-1])
f_d2=np.array([(-1/2),2,(-3/2)])
s_d=np.array([1,-2,1])

spec_f_d=np.fft.fft(f_d)
spec_f_d2=np.fft.fft(f_d2)
spec_s_d=np.fft.fft(s_d)

signal_f_d=thinkdsp.Wave(np.convolve(signal.ys,f_d,'same'),41000)
signal_f_d2=thinkdsp.Wave(np.convolve(signal.ys,f_d2,'same'),41000)
signal_s_d=thinkdsp.Wave(np.convolve(signal.ys,s_d,'same'),41000)
```

```python
thinkplot.preplot(1)
thinkplot.plot(spec_f_d, label='first derivative')
thinkplot.plot(spec_f_d2, label = 'better first derivative')
thinkplot.plot(spec_s_d, label = 'second derivative')
thinkplot.config(loc='best')
```

```python
signal.make_spectrum().plot()
```

```python
signal_f_d.make_spectrum().plot()
```

Certainly the first derivative function has changed the shape of the graph from something that appears to decay by $1/f^2$ to something that decays more slowly. It even appears to decay more slowly than linear (?) but the later regions appear linear. Presumably the densely populated dark areas are due to aliasing, and if you reflected over the aliased triangle you'd see very linear behavior.

```python
signal_f_d2.make_spectrum().plot()
```

I really don't know what to say about this one which is supposed to be the second order approximation of the first derivative. Over the same range as the first signal, amplitude appears to increase by frequency until about 14000 Hz, then decline. Again, it appears that if you were to reflect over the aliased region that looks like a triangle, you'd see the linear decay.

```python
signal_s_d.make_spectrum().plot()
```

This spectrum, that of the signal convolved against the second derivative coefficients, looks like it increases linearly up to a point, then starts to linearly decrease, including into the folded over region. The increasing region sort of matches the expected behavior, since the original signal that decays by 1/f was presumably multiplied against f, and then f again. What would be the result of convolving against [1,-1] twice?

```python
signal_f_d=thinkdsp.Wave(np.convolve(signal_f_d.ys,f_d,'same'),41000)
signal_f_d.make_spectrum().plot()
```

At least this makes some sense! Convolving against [1,-1] twice looks like convolving against [1,-2,1], showing that convolving by the first derivative twice is equivalent to convolving against the second derivative once.

The above graph shows the spectra of the various derivative functions.

5) The gunshot recording I used in chap09.ipynb was clipped, which means that the loudest part gets cut off, which makes the rest of the sound seem louder than it should be, which exaggerates the reverb of the space.

But it looks like there is a better version of this recording at http://www.freesound.org/people/Kleeb/sounds/180960/

Use this recording to replace the one I used and see if it tones down the reverb a little.

Also, apply the same effect to the saxophone recording.

```python
response = thinkdsp.read_wave('180960__kleeb__gunshot.wav')
response = response.segment(start=0.26, duration=5.0)
response.normalize()
response.make_audio()
transfer = response.make_spectrum()
```

```python
violin = thinkdsp.read_wave('../code/92002__jcveliz__violin-origional.wav')
violin.ys = violin.ys[:len(response)]
violin.normalize()
spectrum = violin.make_spectrum()
len(spectrum.hs), len(transfer.hs)
output = (spectrum * transfer).make_wave()
output.normalize()
output.make_audio()
```

Sure enough, the cleaner recording sounds like it has slightly less reverb, but otherwise sounds about the same. I would assume this occurs because the clipped signal introduces frequencies that should not be in the signal, thus producing an output violin sound that has more frequencies (or at least, more of some frequencies) too.

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
parabola=thinkdsp.ParabolicSignal().make_wave().make_spectrum()
parabola.plot()
```

```python
parabola.differentiate()
parabola.plot()
```

The differentiated spectrum appears to be that of a sawtooth wave! Here's the plot of a sawtooth wave to check.

```python
saw=thinkdsp.SawtoothSignal().make_wave().make_spectrum().plot()
```

This relationship can be explained by the differentation operator behaving like a filter where amplitude is proportional to f. If the parabolic signal's original characteristic has amplitude proportional to $f^2$, then applying the differentation filter will give you amplitude proportional to $f$, which is the behavior of the sawtooth signal.
