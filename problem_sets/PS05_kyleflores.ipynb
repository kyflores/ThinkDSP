1) Please finish this exercise that we started in class: One of the examples in chap07.ipynb compares two mixtures of complex sinusoids, one with offset=0 and one with offset=1.  The result is a different looking waveform, because "rotating" each component by the same phase offset does not shift each component by the same amount of time.

I'll start by copying over the example from chap07, then try to rotate it as specified. It should be possible to rotate the entire wave without changing the amplitudes by multiplying by a complex number whose magnitude is 1, and phase angle is not 1, like 0+1j

```python
def synthesize1(amps, freqs, ts):
    components = [thinkdsp.ComplexSinusoid(freq, amp)
                  for amp, freq in zip(amps, freqs)]
    signal = thinkdsp.SumSignal(*components)
    ys = signal.evaluate(ts)
    return ys

amps = np.array([0.6, 0.25, 0.1, 0.05])
freqs = [100, 200, 300, 400]
framerate = 11025

ts = np.linspace(0, 1, framerate)
ys = synthesize1(amps, freqs, ts)

thinkplot.plot(ts[:500], ys[:500].real)
thinkplot.plot(ts[:500], ys[:500].imag)
```

```python
def shift(amps,freqs, offset):
    phases = np.array(freqs) * offset * 2*np.pi
    amp_mod=np.empty(len(freqs),dtype=complex)
    for n in range(len(phases)):
        amp_mod[n]=np.exp(phases[n]*1j)
    return np.multiply(amp_mod,amps)

amps_new=shift(amps,freqs,np.pi)

y=synthesize1(amps_new,freqs,ts)

thinkplot.plot(ts[:500], y[:500].real)
thinkplot.plot(ts[:500], y[:500].imag)
```

One way to shift the waves would be to think of the inputs as being of the form $e^{i2{\pi}ft}$. Multiplication by $e^{i2{\pi}f{\Delta}t}$ shifts the wave by ${\Delta}t$ because multiplying two exponentials adds the exponents.

2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation.

Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
def dft(a):
    """
    input a is a numpy array of values representing
    the time signal to transform

    output h is h[f]=amp
    
    """
    PI2=np.pi*2
    N=len(a)
    
    h=np.empty(N,dtype=np.complex)#Preallocated, because we know exact size.
    ntmp=np.empty(N,dtype=complex)
    
    for k in range(N):
        #Outer assigns dot-sum results to frequency index.
        for n in range(N):
            #Inner loop computes convolution of signal against complex basis.
            ntmp[n]=a[n]*np.exp(-1j*PI2*k*n/N)
        h[k]=np.sum(ntmp)

    return h
```

The implementation of the DFT is astonishingly brief in code. My approach: For every index, compute the correlation/convolution of the signal against a basis sinusoid for lag=0. Real part of the result reflects cosines, imaginary reflects sines. Repeat for basis sinusoids of every frequency up to the size of the input vector (which is the maximum frequency we can expect to sample correctly). Below is a quick test to show that the result is like np.fft.ftt.

```python
import scipy.signal as ssi

a=np.linspace(0,100)
si=np.sin(a)
sq=ssi.square(a)
print("Results correct for sinusoid?",)
print(np.allclose(np.fft.fft(si),dft(si)))
print("\n")
print("Results correct for square wave?",)
print(np.allclose(np.fft.fft(sq),dft(sq)))
```

I'm fairly confident of my result because it passes with both a very simple sin wave and a more interesting square wave. Numpy's allclose was used to check equality within default tolerances (1e-5 relative, 1e-8 absolute). This tolerance is needed to check equality within small floating point error tolerances which I expect can accrue very easily in this implementation due to the large amount of multiplies and reduces. 

This could probably be sped up a bit by vectorizing the inner loop to use the numpy optimized dot product.

3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.

Based on the wiki page, it sounds like I need to break up the input into the smallest practical size by splitting the input up at each stage in a vector of even indicies and of odd indicies until the vector only contains one item, at which point the spectrum of the single point is itself. I think this occurs because the summation in the DFT formulation goes away and the entire exponential component becomes equal to 1. In this case, $X_k=x_n*1$

```python
def fft(a):
    """
    input length must be a power of two.

    input: time signal numpy array
    output: h[f]=amp numpy array
    """
    PI2=np.pi*2
    N=len(a)

    if len(a)<=1:
        #Base case reached
        return a
    else:
        tw=np.empty(N,dtype=complex) #tw are the twiddle factors.
        for k in range(N):
            tw[k]=np.exp(-1j*PI2*k/N)

        h_e=fft(a[::2]) #All even indicies
        h_o=fft(a[1::2]) #All odd indicies
        H=np.empty(N,dtype=np.complex)
        for k in range(N):
            i=(k%(N/2)) #For indexing the half size arrays
            H[k]=h_e[i]+tw[k]*h_o[i]

        return H
```

```python
a=np.linspace(0,100,128)
si=np.sin(a)
sq=ssi.square(a)
print("Did my implementation produce the correct results for a sinusoid? ",)
print(np.allclose(np.fft.fft(si),fft(si)))
print("\n")
print("Did my implementation produce the correct results for a square wave?",)
print(np.allclose(np.fft.fft(sq),fft(sq)))
```

My fft passes the same test results as before. Presumably the numpy fft has some way of automatically dealing with lengths that aren't a power of 2.

Incidentally, the entire recursion sequence can be skipped by bit reversing the indicies of the input vector, that is, look at the binary representation of the index and reverse it. 110 becomes 011, 001 becomes 100, etc. Then, the sequence can be propagated forwards with addition of the correct pairs of indicies, and multiplication by the twiddle factors. This method seems better suited for multi-threaded applications and probably grows the stack less.

4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

I'll be reusing a lot of code from PS03, as well as pasting in the functions to benchmark.

```python
import scipy

def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

def fft_convolve(signal, window):
    fft_signal = np.fft.fft(signal)
    fft_window = np.fft.fft(window)
    return np.fft.ifft(fft_signal * fft_window)

def zero_pad(array, n):
    res = np.zeros(n)
    res[:len(array)] = array
    return res

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ns = 2 ** np.arange(4, 12)
results1 = []
results2 = []
results3 = []
for N in ns:
    ts = (0.5 + np.arange(N)) / N
    freqs = (0.5 + np.arange(N)) / 2
    ys = noise.ys[:N]
    test_wave=thinkdsp.Wave(ys,11025)
    result1 = %timeit -o autocorr(test_wave)
    results1.append(result1)
    result2 = %timeit -o fft_autocorr(ys)
    results2.append(result2)
    result3 = %timeit -o np.correlate(ys,ys)
    results3.append(result3)

bests1 = [result.best for result in results1]
bests2 = [result.best for result in results2]
bests3 = [result.best for result in results3]
```
