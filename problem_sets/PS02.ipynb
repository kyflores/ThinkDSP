2) "Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly."

Writing this class will involve inheriting from Chirp and and overloading evaluate, and probably also \_evalute. I've included the Sawtooth class again for comparison, and wrote this class by modifying the \_evaluate function of chirp.

```python
class Sawtooth(thinkdsp.Sinusoid):
    """
    A sawtooth wave defined by frequency, amplitude, and a phase offset
    """
    def evaluate(self,ts):
        """
        Evaluates a sawtooth signal over a specific time series.
        """
        self.cycles=self.freq*ts + self.offset/thinkdsp.PI2
        self.frac, self._ = np.modf(self.cycles)
        self.ys = self.frac - 0.5 #Removing the numpy.abs(self.frac - 0.5) here was the only change I made.
        self.ys = thinkdsp.normalize(thinkdsp.unbias(self.ys), self.amp)
        return self.ys

class SawtoothChirp(thinkdsp.Chirp):
    """
    A sawtooth (rather than sinusoidal) chirp function.
    """
    def _evaluate(self, ts, freqs):
        self.dts = np.diff(ts)
        self.dps = thinkdsp.PI2 * freqs * self.dts
        self.phases = np.cumsum(self.dps)
        self.phases = np.insert(self.phases, 0, 0)
        self.frac, self._ = np.modf(self.phases/thinkdsp.PI2)
        self.ys=self.frac-0.5
        ys = thinkdsp.normalize(thinkdsp.unbias(self.ys),self.amp)
        return ys

saw_chirp=SawtoothChirp()
saw_chirp.make_wave().make_audio()
```

The \_evaluate function was modified by using the phases generated for a chirp, then evaluating them for ys in a sawtooth like function. Now, I'll make the spectrograph of the sawtooth chirp.

```python
saw_chirp.make_wave().make_spectrogram(seg_length=128).plot()
```

```python
saw_chirp.make_wave().make_spectrogram(seg_length=2048).plot()
```

The spectrogram for the sawtooth chirp is a group of sloped lines because there should be a fixed ratio between the fundamental and its harmonics. There appears to be the possibility of aliasing with a short seg_length, so I presented a graph with a longer seg_length as well.

3)  "In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length."

The trombone's glissando as described would be most like an exponential chirp. Pitch and intervals are perceptual constructs that are dependent on a ratio between two frequencies. An octave for instance, is always a frequency ratio of 2:1, but the difference in frequency will vary. For instance, 220Hz and 440Hz are perceived as being an octave apart, as well as 440Hz and 880Hz. In order to increase pitch linearly, you would need to cross the first 220Hz in the same time you cross the next 440Hz.

```python
def trombone_gliss_expo(lowf,highf):
    trombone_gliss_up=thinkdsp.ExpoChirp(lowf,highf).make_wave()
    trombone_gliss_dn=thinkdsp.ExpoChirp(highf,lowf).make_wave()
    return (trombone_gliss_up|trombone_gliss_dn) #Note to self: | is __or__ and concats waves.

def trombone_gliss_lin(lowf,highf):
    trombone_gliss_up=thinkdsp.Chirp(lowf,highf).make_wave()
    trombone_gliss_dn=thinkdsp.Chirp(highf,lowf).make_wave()
    return (trombone_gliss_up|trombone_gliss_dn) #Note to self: | is __or__ and concats waves.

trombone_gliss_expo(262,349).make_audio()
```

```python
trombone_gliss_lin(262,349).make_audio()
```

The functions defined above produce the gliss with both a linear and exponential chirp. To my ear, it is quite difficult to tell the difference for the relatively small range of frequencies swept. I can really hear the difference if make the sweep wider, like 100 Hz to 5000 Hz.
