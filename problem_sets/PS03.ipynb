2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

Plotting the autocorrelation should show linearly decreasing correlation. For a small lag, the values being considered are close in time and thus more correlated than values far in time. To plot the autocorrelation, I'll grab the autocorrelation function from chap05 and build around it.

```python
def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

(lags,corrs)=autocorr(wave)
thinkplot.plot(lags,corrs)
```

Sure enough, the correlation decreases with increasing lag. The region that increases again at the end suggests some sort of periodicity, but the magnitude of the two maxima are very different, so the periodicity is probably just a coincidence.

4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

analyze1 and analyze2 imported from dct. I'm using the PS03 starter code to produce the test results.


```python
import scipy.fftpack
import dct

signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ns = 2 ** np.arange(4, 12)
results1 = []
results2 = []
results3 = []
for N in ns:
    ts = (0.5 + np.arange(N)) / N
    freqs = (0.5 + np.arange(N)) / 2
    ys = noise.ys[:N]
    result1 = %timeit -o dct.analyze1(ys, freqs, ts)
    results1.append(result1)

bests1 = [result.best for result in results1]
    
for N in ns:
    ts = (0.5 + np.arange(N)) / N
    freqs = (0.5 + np.arange(N)) / 2
    ys = noise.ys[:N]
    result2 = %timeit -o dct.analyze2(ys, freqs, ts)
    results2.append(result2)

bests2 = [result.best for result in results2]

for N in ns:
    ts = (0.5 + np.arange(N)) / N
    freqs = (0.5 + np.arange(N)) / 2
    ys = noise.ys[:N]
    result3 = %timeit -o scipy.fftpack.dct(ys)
    results3.append(result3)

bests3 = [result.best for result in results3]
```

```python
thinkplot.plot(ns, bests1)
#thinkplot.config(xscale='log', yscale='log')
```

```python
x = np.log(ns)
y = np.log(bests1)
t = scipy.stats.linregress(x,y)
t
```

The first algorithm grows exponentially with input size.

5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:
Break a long signal into segments.
Compute the DCT of each segment.
Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.
Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
def dct_comp(wave,s_l,thresh):
    """
    wave is a thinkdsp wave object we want to compress
    s_l is the segment length
    """
    assert type(wave)==thinkdsp.Wave
    ys_all=np.array_split(wave.ys,s_l)
    dcts=[]
    del_ind=[]
    for a in range(len(ys_all)):
        dcts.append(scipy.fftpack.dct(ys_all[a].astype(np.float64),norm='ortho'))
        for i in xrange(len(dcts[a])):
            if (abs(dcts[a][i]) < thresh):
                dcts[a][i]=0 #Think of this as equivalent to deleting the value, but without all the indexing annoyances.
        dcts[a]=scipy.fftpack.idct(dcts[a],norm='ortho')

    ys_o=np.concatenate(dcts)
    return thinkdsp.Wave(ys_o,wave.framerate)
```

```python
amp_thresh=10000

wave=thinkdsp.read_wave('../code/100475__iluppai__saxophone-weep.wav')
comp=dct_comp(wave,128,50000)
```

```python
size_in=sum(scipy.fftpack.dct(abs(wave.ys.astype(np.float64))))
size_out=sum(scipy.fftpack.dct(abs(comp.ys.astype(np.float64))))
size_out/size_in
```
