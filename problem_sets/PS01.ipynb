Next, I'll write a class that implements a Sawtooth signal class and a means for evaluating it. Recognizing that a sawtooth wave is similar in behavior to a triangle wave (at least in implementation) I have implemented the sawtooth wave with a minor modification to the triangle wave class definition.

```python
class Sawtooth(thinkdsp.Sinusoid):
    """
    A sawtooth wave defined by frequency, amplitude, and a phase offset
    """
    def evaluate(self,ts):
        """
        Evaluates a sawtooth signal over a specific time series.
        """
        self.cycles=self.freq*ts + self.offset/thinkdsp.PI2
        self.frac, self._ = np.modf(self.cycles)
        self.ys = self.frac - 0.5 #Removing the numpy.abs(self.frac - 0.5) here was the only change I made.
        self.ys = thinkdsp.normalize(thinkdsp.unbias(self.ys), self.amp)
        return self.ys
```

```python
saw_sig=Sawtooth()
saw_sig.make_wave(duration=0.02).plot()
```

```python
saw_sig.make_wave(duration=0.1).make_spectrum().plot()
```

The spectrum produced contains all integer multiples of the fundamental. Looking at the approximate y axis values, it looks like the amplitude drops of by a factor of $1/f$ as the frequency increases. After seeing this behavior, another interesting way to implement the sawtooth wave might have been to sum a series of signals of decaying amplitude. The following cell shows the result of this sum approach with 5 harmonics, just for curiosity's sake.

```python
sigs=[]
for n in range(5):
    sigs.append(thinkdsp.SinSignal(freq=440*(n+1),amp=100/(n+1)))
    
sawtooth=thinkdsp.SumSignal(*sigs)
sawtooth.make_wave(duration=0.01).plot()
```

Relative to the square wave, the sawtooth wave contains all harmonics instead of just odd harmonics. The harmonic amplitudes drop off at $1/f$ for both the square and sawtooth. Relative to the triangle wave, the sawtooth wave contains more harmonics (not just odd) and drops off more slowly. It appears the triangle wave drops off at $1/f^2$

Next I'll try to make a wave that has both even and odd harmonics that drop off by $1/f^2$. I'll choose to begin by synthesizing this wave from its definition, similar to how I make the sawtooth from a sum of signals earlier.

```python
sigs=[]
for n in range(7):
    sigs.append(thinkdsp.SinSignal(freq=440*(n+1),amp=100/(n+1)**2))
    
sawtooth=thinkdsp.SumSignal(*sigs)
sawtooth.make_wave(duration=0.01).plot()
sawtooth.make_wave().make_audio()
```

I certainly couldn't say for sure what mathematical function this is, qualitatively it looks like a rounder sawtooth wave. I'll try run the same piece of code again but this time include a phase offset to each individual component signal which is equal to the square of the multiple it is of the fundamental.

```python
sigs=[]
for n in range(7):
    sigs.append(thinkdsp.SinSignal(freq=440*(n+1),amp=100/(n+1)**2,offset=n**2))
    
sawtooth=thinkdsp.SumSignal(*sigs)
sawtooth.make_wave(duration=0.01).plot()
sawtooth.make_wave().make_audio()
```

Obviously this wave looks very different but unsurprisingly sounds exactly the same to our ear. All waves with the same harmonic content should sound nearly if not exactly the same to our ear. I don't believe humans are, on average, able to really tell the difference when a phase shift occurs.
